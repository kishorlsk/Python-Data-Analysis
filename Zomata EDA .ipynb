```python
country_df = pd.read_excel('Country-Code.xlsx')
country_df.head()
```

```python
df.columns
```

```python
#country code is matching
```

```python
final_df = pd.merge(df,country_df, on='Country Code', how='left')
final_df.head()
```

```python
#checking data types
final_df.dtypes
```

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
country_val = final_df['Country'].value_counts().values
country_val
```

```python
#top 3 country uses Zomato

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.1f%%')
```

OBSERVATION:

1.Zomato maximum records or transaction are from India after that United States and United Kingdom

```python
final_df.columns
```

```python
rating_df = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
rating_df
```

#OBSERVATION

1.when Rating is between 4.5 to 4.9 ----> Excellent

2.when Rating is between 4.0 to 4.4 ----> Very Good

3.when Rating is between 3.5 to 3.9 ----> Good

4.when Rating is between 2.5 to 3.4 ----> Average

5.when Rating is between 1.8 to 2.4 ----> Poor

5.when Rating is 0 ----> Not rated

```python
rating_df.head()
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=rating_df)
plt.show()
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=rating_df,hue='Rating color',palette=['blue','red','orange','yellow','lightgreen','darkgreen'])
```

#OBSERVATION

  1.Not rated count is very high
  
  2.Maximum number of rating is between 2.9 to 3.9
  
  3.Note distribution looks like normal distribution
