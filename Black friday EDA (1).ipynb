```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing train Dataset
df_train= pd.read_csv('black_train.csv')
df_train.head()
```

```python
#importing test Dataset
df_test = pd.read_csv('black_test.csv')
df_test.head()
```

```python
#Merge both train and test dataset

df = df_train.append(df_test)
df.head()
```

```python
#Handling categorical feature gender to numerical feture
```

```python
df['Gender'] = df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
##Handling categorical feature Age
```

```python
df['Age'] =  df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
##Handling categorical feature City_Category
```

```python
df_city = pd.get_dummies(df['City_Category'],drop_first=True)
df_city
```

```python
df =  pd.concat([df,df_city],axis=1)
```
